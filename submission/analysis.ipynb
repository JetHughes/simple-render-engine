```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python

data = pd.read_csv("data.csv")
data.describe()
```

```python
x = range(len(data))
y = data

plt.plot(x, y)
plt.ylim(1, 5) 
plt.xticks([])
plt.xlabel('Time')
plt.ylabel('Ms per frame')
plt.show()
```

```python
plt.boxplot(data, vert=False)
plt.xlabel("ms/frame")
plt.yticks([])
plt.title("Box plot of ms/frame of render engine")
plt.show()
```
